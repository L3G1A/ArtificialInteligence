```python
from random import random
import math



counter = 0

for i in range(0,10**3):
    cord = [random(), random()]
    if((cord[0]**2 + cord[1]**2) < 1):
        counter += 1
        

pi3 =  4 * (counter/10**3)



counter = 0

for i in range(0,10**4):
    cord = [random(), random()]
    if((cord[0]**2 + cord[1]**2) < 1):
        counter += 1
        

pi4 =  4 * (counter/10**4)


counter = 0

for i in range(0,10**5):
    cord = [random(), random()]
    if((cord[0]**2 + cord[1]**2) < 1):
        counter += 1
        

pi5 =  4 * (counter/10**5)

counter = 0

for i in range(0,10**6):
    cord = [random(), random()]
    if((cord[0]**2 + cord[1]**2) < 1):
        counter += 1
        

pi6 =  4 * (counter/10**6)




print("n = 10 ^ 3 pi = " , pi3 , " error = " , abs((math.pi - pi3)/math.pi) * 100 , "%")
print("n = 10 ^ 4 pi = " , pi4 , " error = " , abs((math.pi - pi4)/math.pi) * 100 , " %")
print("n = 10 ^ 5 pi = " , pi5 , " error = " , abs((math.pi - pi5)/math.pi) * 100 , " %")
print("n = 10 ^ 6 pi = " , pi6 , " error = " , abs((math.pi - pi6)/math.pi) * 100 , " %")
```

```python
import matplotlib.pyplot as plt
from random import random

ys = []
xs = []
colors = []
edgecolors = []
for i in range(0,10**4):
    cord = [random(), random()]

    if((cord[0]**2 + cord[1]**2) < 1):
        ys.append(cord[1])
        xs.append(cord[0])
        colors.append('blue')
        edgecolors.append("white")
    else:
        ys.append(cord[1])
        xs.append(cord[0])
        colors.append('red')
        edgecolors.append("white")

plt.scatter(ys, xs, color=colors,  edgecolor=edgecolors)
        
plt.show()

```
